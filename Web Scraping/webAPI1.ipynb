# Web APIs

- Application programming Interface -> set of routines running over Web Server
- Connects Application and Database
- Specify Data with URL
- Retrieve Data and Parse it
- API may return data in JSON,XML data format

### JSON
- JavaScript Object Notation
- Dictionary like Data Structure : key value pair
- keys are strings,values can be anything
- easy to read and write

### XML
- Extensible Markup language
- Structure similar to XML
- Custom Tags

```python
from urllib.request import urlopen
```

```python
api_url = "https://samples.openweathermap.org/data/2.5/weather?q=london&appid=b6907d289e10d714a6e88b30761fae22"
```

```python
url_result = urlopen(api_url)
print(url_result)
```

```python
data = url_result.read()
print(data)
```

```python
print(type(data))
```

**Convert this data into JSON Object**

```python
import json   # python library like dictionary DS
json_data = json.loads(data)
print(json_data)
print(type(json_data))
```

```python
json_data['coord']
```

```python
json_data['name']
```

```python
# converting json into string
json_string = json.dumps(json_data) 
print(json_string)
print(type(json_string))
```

### Beautiful Soup
Beautiful Soup is a Python library for pulling data out of HTML,and XML files.It works with your favourite parser to provide idiomatic ways of navigating,searching, and modifying the parse tree.It commonly saves programmers hours or days of work.

```python
!pip install bs4
```

```python
!pip install html5lib
```

```python
import bs4
import requests
```

```python
url = "https://www.passiton.com/inspirational-quotes?page=2"
```

```python
res = requests.get(url)
```

```python
res.content
```

```python
# HTML parser
soup = bs4.BeautifulSoup(res.content)
```

```python
soup.findAll?
```

```python
soup
```

```python
sec_element = soup.findAll('section')
```

```python
sec_element[1]
```

```python
section = sec_element[1]
```

```python
section.a
```

```python
section.a
```

```python
section.img
```

```python
section.img.attrs
```

```python
section.img.attrs['src']
```

```python
with open ('inspirational.jpg','wb') as f:
    image_url = section.img.attrs['src']
    res = requests.get(image_url)
    
    f.write(res.content)
```
