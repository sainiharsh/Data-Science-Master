```python
import requests
```

```python
url = "http://www.icndb.com/"
```

```python
r = requests.get(url)
```

```python
r.content
```

```python
import bs4
```

```python
new_r = bs4.BeautifulSoup(r.content)
```

```python
new_r
```

```python
new_r.em
```

```python
new_r.findAll("em",{})
```

```python
joke = new_r.findAll("em")
```

```python
print(joke[0].text)
```

```python
filename = 'joke.csv'
with open(filename,'w',encoding = 'utf-8') as f:
    
    f.write(joke[0].text)
```
