```python
soup = bs4.BeautifulSoup(res.content)
```

```python
div_ele = soup.findAll('img')
```

```python
div_ele
```

```python
i = div_ele[1]
```

```python
i.attrs['src']
```

```python
for i,image in enumerate(div_ele):
    with open("YMCA{}.jpg".format(i),'wb') as f:
        img_url = image.attrs['src']
        res = requests.get(img_url)
        
        f.write(res.content)
```
